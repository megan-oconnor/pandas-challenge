# PyCity Schools Analysis

```python
# import packages
import pandas as pd
from pathlib import Path
```

```python
# filepaths
school_file = Path('Resources/schools_complete.csv')
student_file = Path('Resources/students_complete.csv')

# load files
school_df = pd.read_csv(school_file)
student_df = pd.read_csv(student_file)

# combine data into one dataset
df = pd.merge(student_df, school_df, how='left', on=['school_name', 'school_name'])
df.head()
```

## District Summary

```python
# total number of unique schools
n_schools = df['school_name'].nunique()
n_schools
```

```python
# total students
n_students = df['Student ID'].nunique()
n_students
```

```python
# total budget
tt_budget = school_df['budget'].sum()
tt_budget
```

```python
# average math score
avg_math = df['math_score'].mean()
avg_math
```

```python
# average reading score
avg_read = df['reading_score'].mean()
avg_read
```

```python
# % passing math
pass_math = df.loc[df['math_score']>69,'math_score'].count()
prt_pass_math = 100*(pass_math/n_students)
prt_pass_math
```

```python
# % passing reading
pass_read = df.loc[df['reading_score']>69,'reading_score'].count()
prt_pass_read = 100*(pass_math/n_students)
prt_pass_read
```

```python
# % passing math and reading
pass_mathread = df.loc[(df['math_score']>69) & (df['reading_score']>69),'math_score'].count()
prt_pass_mathread = 100*(pass_mathread/n_students)
prt_pass_mathread
```

```python
# summary stats
district_summary = pd.DataFrame([{'Total Schools': n_schools,'Total Students': n_students, 'Total Budget': tt_budget,
                                  'Average Math Score': avg_math,'Average Reading Score': avg_read,
                                  '% Passing Math': prt_pass_math,'% Passing Reading': prt_pass_read,
                                  '% Overall Passing': prt_pass_mathread}])

# formatting
district_summary["Total Students"] = district_summary["Total Students"].astype(float).map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].astype(float).map("${:,.2f}".format)

# display table
district_summary
```

## School Summary

```python
# each schools type
school_type = school_df.set_index(['school_name'])['type']
school_type
```

```python
# total students per school
school_size = school_df.set_index(['school_name'])['size']
school_size
```

```python
# budget per school
school_budget = school_df.set_index(['school_name'])['budget']
school_budget

# budget per capita
school_budget_perCap = school_budget/school_size
school_budget_perCap
```

```python
# group df by school
grouped = df.groupby('school_name')

# average math score per school
avg_math_by = grouped['math_score'].mean()
avg_math_by
```

```python
# average reading score per school
avg_read_by = grouped['reading_score'].mean()
avg_read_by
```

```python
# % passing math per school
pass_math = df.loc[df['math_score']>69,:]
group_pass_math = pass_math.groupby('school_name').count()

prt_pass_math_school = 100*(group_pass_math['math_score']/school_size)
prt_pass_math_school
```

```python
# % passing reading per school
pass_read = df.loc[df['reading_score']>69,:]
group_pass_read = pass_read.groupby('school_name').count()

prt_pass_read_school = 100*(group_pass_read['reading_score']/school_size)
prt_pass_read_school
```

```python
# % passing math and reading
pass_mathread = df.loc[(df['reading_score']>69) & (df['math_score']>69),:]
group_pass_mathread = pass_mathread.groupby('school_name').count()

prt_pass_mathread_school = 100*(group_pass_mathread['Student ID']/school_size)
prt_pass_mathread_school
```

```python
# schools summary
school_sum = pd.DataFrame({'School Type': school_type,'Total Students': school_size,
                               'Total School Budget': school_budget,'Per Student Budget': school_budget_perCap,
                               'Average Math Score': avg_math_by,'Average Reading Score': avg_read_by,
                               '% Passing Math': prt_pass_math_school,'% Passing Reading': prt_pass_read_school,
                               '% Overall Passing': prt_pass_mathread_school})

# create a copy to edit formats
school_summary = school_sum.copy()

# formatting
school_summary["Total Students"] = school_summary["Total Students"].map("{:,}".format)
school_summary["Total School Budget"] = school_summary["Total School Budget"].map("${:,.2f}".format)
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].map("${:,.2f}".format)

# display DataFrame
school_summary
```

## Highest-Performing Schools (by % Overall Passing)

```python
#sort school summary from highest-performing schools to lowest-performing schools
sort_school_overall= school_summary.sort_values('% Overall Passing', ascending = False)

#show the 5 highest performing schools
sort_school_overall.head(5)
```

## Math Scores by Grade

```python
# separate by grade
ninth = df[df['grade'] == '9th']
tenth = df[df['grade'] == '10th']
eleventh = df[df['grade'] == '11th']
twelfth = df[df['grade'] == '12th']

# average math score for students by grade at each school
# group each dataset by school and get the mean of the math score
nin_avg_math = ninth.groupby('school_name')['math_score'].mean()
ten_avg_math = tenth.groupby('school_name')['math_score'].mean()
ele_avg_math = eleventh.groupby('school_name')['math_score'].mean()
twe_avg_math = twelfth.groupby('school_name')['math_score'].mean()

# combine into a single DataFrame
math_scores_by_grade = pd.DataFrame({'9th': nin_avg_math,'10th': ten_avg_math,'11th': ele_avg_math,'12th': twe_avg_math})

# minor data wrangling
math_scores_by_grade.index.name = None

# display the DataFrame
math_scores_by_grade
```

```python
# average reading score for students by grade at each school
# group each dataset by school and get the mean of the math score
nin_avg_read = ninth.groupby('school_name')['reading_score'].mean()
ten_avg_read = tenth.groupby('school_name')['reading_score'].mean()
ele_avg_read = eleventh.groupby('school_name')['reading_score'].mean()
twe_avg_read = twelfth.groupby('school_name')['reading_score'].mean()

# combine into a single DataFrame
read_scores_by_grade = pd.DataFrame({'9th': nin_avg_read,'10th': ten_avg_read,'11th': ele_avg_read,'12th': twe_avg_read})

# minor data wrangling
read_scores_by_grade.index.name = None

# display the DataFrame
read_scores_by_grade
```

## Scores by School Spending

```python
# bins for spending
bins = [0,585,630,645,680]
labels = ['<$585','$585-630','$630-645','$645-680']
```

```python
# create a copy of the original DataFrame
school_spend = school_sum.copy()
```

```python
# categorize the per capita budget with the labels
school_spend['Spending Ranges (Per Student)'] = pd.cut(school_spend['Per Student Budget'], bins, labels=labels, include_lowest=True)

# format table values
school_spend["Total Students"] = school_spend["Total Students"].map("{:,}".format)
school_spend["Total School Budget"] = school_spend["Total School Budget"].map("${:,.2f}".format)
school_spend["Per Student Budget"] = school_spend["Per Student Budget"].map("${:,.2f}".format)

school_spend
```

```python
# calculate averages for scores by spending range
spending_math_scores = school_spend.groupby(['Spending Ranges (Per Student)'])['Average Math Score'].mean()
spending_read_scores = school_spend.groupby(['Spending Ranges (Per Student)'])['Average Reading Score'].mean()
spending_passing_math = school_spend.groupby(['Spending Ranges (Per Student)'])['% Passing Math'].mean()
spending_passing_read = school_spend.groupby(['Spending Ranges (Per Student)'])['% Passing Reading'].mean()
spending_passing_overall = school_spend.groupby(['Spending Ranges (Per Student)'])['% Overall Passing'].mean()
```

```python
# summary table
spend_summary = pd.DataFrame({'Average Math Score': spending_math_scores, 'Average Reading Score': spending_read_scores,
                               '% Passing Math': spending_passing_math, '% Passing Reading': spending_passing_read,
                               '% Overall Passing': spending_passing_overall})

# display table
spend_summary
```

## Scores by School Size

```python
# bins for school size
size_bins = [0, 1000, 2000, 5000]
size_labels = ['Small (<1000)', 'Medium (1000-2000)', 'Large (2000-5000)']
```

```python
# create a copy of the original DataFrame
school_size = school_sum.copy()
```

```python
# categorize the school sizes
school_size['School Size'] = pd.cut(school_size['Total Students'], size_bins, labels=size_labels, include_lowest=True)

# format table values
school_size["Total Students"] = school_size["Total Students"].map("{:,}".format)
school_size["Total School Budget"] = school_size["Total School Budget"].map("${:,.2f}".format)
school_size["Per Student Budget"] = school_size["Per Student Budget"].map("${:,.2f}".format)

school_size
```

```python
# calculate averages for scores by spending range
size_math_scores = school_size.groupby(['School Size'])['Average Math Score'].mean()
size_read_scores = school_size.groupby(['School Size'])['Average Reading Score'].mean()
size_passing_math = school_size.groupby(['School Size'])['% Passing Math'].mean()
size_passing_read = school_size.groupby(['School Size'])['% Passing Reading'].mean()
size_passing_overall = school_size.groupby(['School Size'])['% Overall Passing'].mean()
```

```python
# summary table
size_summary = pd.DataFrame({'Average Math Score': size_math_scores, 'Average Reading Score': size_read_scores,
                             '% Passing Math': size_passing_math, '% Passing Reading': size_passing_read,
                             '% Overall Passing': size_passing_overall})

# display table
size_summary
```
